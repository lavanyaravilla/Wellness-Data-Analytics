```python
import glob
import os
import pandas as pd

# Define the folder path containing your consolidated scraped CSVs
# Using a raw string (r"...") prevents issues with backward slashes in Windows paths
folder_path = r"C:\Users\lavan\OneDrive\Desktop\scrapping\output\FitnessRawDate_justdial"

# Create a search pattern for all .csv files in that directory
csv_pattern = os.path.join(folder_path, "*.csv")

# Get a list of all CSV file paths matching the pattern
csv_files = glob.glob(csv_pattern)

print(f"Found {len(csv_files)} CSV files to merge.")

# Read each CSV file into a DataFrame and store them in a list
# This automatically handles matching column headers like name, address, rating, etc.
df_list = []
for file in csv_files:
    try:
        # Using encoding='utf-8' or 'latin1' handles hidden special characters often found in web-scraped data
        df = pd.read_csv(file, encoding="utf-8")
        df_list.append(df)
        print(f"Successfully loaded: {os.path.basename(file)}")
    except Exception as e:
        print(f"Error reading {os.path.basename(file)}: {e}")

# Merge (concatenate) all dataframes vertically
if df_list:
    merged_df = pd.concat(df_list, ignore_index=True)

    # Optional: Remove exact duplicate rows if the same center was scraped in multiple files
    merged_df.drop_duplicates(inplace=True)

    # Save the consolidated file back to the folder
    output_file_path = os.path.join(folder_path, "all_justdial_consolidated.csv")
    merged_df.to_csv(output_file_path, index=False, encoding="utf-8")

    print("\nMerge complete!")
    print(f"Total rows in merged file: {len(merged_df)}")
    print(f"Saved consolidated data to: {output_file_path}")
else:
    print("No CSV files found or loaded.")
```
