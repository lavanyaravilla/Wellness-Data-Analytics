```python
# Cell 2: Import libraries
import requests
from bs4 import BeautifulSoup
```

```python
# Cell 3: Fetch the Justdial page
url = "https://www.justdial.com/Bangalore"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
response = requests.get(url, headers=headers)

# Check status
print("Status Code:", response.status_code)
```

```python
# Cell 3: Define the links
urls = [
    "https://www.justdial.com/Bangalore/Gyms",
    "https://www.justdial.com/Bangalore/Fitness-Studio",
    "https://www.justdial.com/Bangalore/Yoga-Centres",
    "https://www.justdial.com/Bangalore/dance-zumba-studio",
    "https://www.justdial.com/Bangalore/Crossfit-Gyms",
    "https://www.justdial.com/Bangalore/Calisthenic-Gyms",
    "https://www.justdial.com/Bangalore/Pilates-Studios",
    "https://www.justdial.com/Bangalore/Karate-Dojo",
    "https://www.justdial.com/Bangalore/Taekwondo-Academy",
    "https://www.justdial.com/Bangalore/Kickboxing-Studios"
]

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
```

```python


for url in urls:
    print(f"\nScraping: {url}")
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    listings = soup.find_all("div", class_="resultbox")
    for box in listings:
        name = box.find("span", class_="resultbox_title_anchor")
        address = box.find("div", class_="locatcity")
        rating = box.find("li", class_="resultbox_totalrate")
        reviews = box.find("li", class_="resultbox_countrate")
        phone_tag = box.find("span", class_="callcontent")
        phone = phone_tag.get_text(strip=True) if phone_tag else None



        print("Fitness Centre Name:", name.text.strip() if name else None)
        print("Address:", address.text.strip() if address else None)
        print("Rating:", rating.text.strip() if rating else None)
        print("Number of Reviews:", reviews.text.strip() if reviews else None)
        print("Phone Number:", phone if phone else None)
        print("-" * 40)
```
